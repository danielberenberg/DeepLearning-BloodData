```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
path1 = '../archive/final/c3d/training.log'
path2 = '../archive/new/c3d/training.log'
```

```python
path1df = pd.read_csv(path1) #bw
path2df = pd.read_csv(path2) #color
losses1 = list(path1df['loss'])
losses2 = list(path2df['loss'])

path3df = pd.read_csv("../archive/new/lrcn_bw/training.log")
path4df = pd.read_csv("../archive/new/lrcn_color/training.log")

losses3 = list(path3df['loss'])
losses4 = list(path4df['loss'])

path5df = pd.read_csv("../archive/new/cnn_3d_small/training.log")
losses5 = list(path5df["loss"])
```

```python
fig, axarr = plt.subplots(1,3, figsize=(6, 4), sharey=True, sharex=True)
plt.suptitle('Training loss comparison', y=1.05)
axarr[0].plot(losses1, label="BW")
axarr[0].plot(losses2, label="RGB")
axarr[0].set_title("C3D")
#ax = plt.gca()
#plt.locator_params(numbins=10, axis='y')
axarr[0].set_ylabel('Loss')
axarr[0].set_xlabel('Epochs')

#axarr[2].plot(losses3, label="BW")
axarr[1].plot(losses5, label='RGB', color='C1')
axarr[1].set_xlabel("Epochs")
axarr[1].set_title("Small 3D CNN")

#red_patch = mpatches.Patch(color='red', label='loss')
#blue_patch = mpatches.Patch(color='blue', label='validation loss')
#plt.legend(handles=[red_patch, blue_patch])

axarr[2].plot(losses3, label="BW")
axarr[2].plot(losses4, label="RGB")
axarr[2].set_xlabel("Epochs")
axarr[2].set_title("LRCN")


axarr[0].legend()
plt.tight_layout()

plt.savefig('losses.png')
plt.show()
```
